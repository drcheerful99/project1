```python
# Total Traveler Throughput 2020

rcParams['figure.figsize'] = 14,12

date = tsa['Date']
travelers_2020 = tsa['Total Traveler Throughput 2020']
plt.plot(date,travelers_2020, color='r', linewidth=1, marker='.', markersize=5, label='Travelers 2020', linestyle='-')

plt.title('Traveler Numbers Throughout 2020',fontsize = 24)
plt.xlabel('Date',fontsize = 14)
plt.ylabel('Travelers Number',fontsize = 14)

# plt.xlim()
# plt.xticks(date[::])

# plt.ylim()
# plt.yticks([])


plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# comparing 2020 to 2019
from matplotlib import rcParams
rcParams['figure.figsize'] = 14,12

date = tsa['Date']
travelers_2020 = tsa['Total Traveler Throughput 2020']
plt.plot(date,travelers_2020, color='r', marker='.', markersize=5, label='Travelers 2020', linestyle='-')

travelers_2019 = tsa['Total Traveler Throughput 2019']
plt.plot(date,travelers_2019, color='g', linewidth=1, marker='.', markersize=5, label='Travelers 2019', linestyle='-')

plt.title('Traveler Numbers Throughout 2019 VS. 2020',fontsize = 24)
plt.xlabel('Date',fontsize = 14)
plt.ylabel('Travelers Number',fontsize = 14)

# plt.xlim()
# plt.xticks(date[])

# plt.ylim()
# plt.yticks([])

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
