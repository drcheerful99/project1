```python
# Total Traveler Throughput 2020

date = tsa['Date']
travelers_2020 = tsa['Total Traveler Throughput 2020']
plt.plot(date,travelers_2020, label='Travelers 2020', linestyle='-')

# plt.figure(figsize=(5,3), dpi=200)
plt.title('Traveler Numbers Throughout 2020',fontsize = 14)
plt.xlabel('Date',fontsize = 10)
plt.ylabel('Travelers Number',fontsize = 10)
#plt.xlim(1,)
#plt.ylim()
#plt.xticks(['Date'])
#plt.yticks(['Total Traveler Throughput 2020'])

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# comparing 2020 to 2019

date = tsa['Date']
travelers_2020 = tsa['Total Traveler Throughput 2020']
plt.plot(date,travelers_2020, label='Travelers 2020', linestyle='-')

travelers_2019 = tsa['Total Traveler Throughput 2019']
plt.plot(date,travelers_2019, label='Travelers 2019', linestyle='-')

# plt.figure(figsize=(5,3), dpi=200)
plt.title('Traveler Numbers Throughout 2019 comparing to 2020',fontsize = 14)
plt.xlabel('Date',fontsize = 10)
plt.ylabel('Travelers Number',fontsize = 10)
#plt.xlim(1,)
#plt.ylim()
#plt.xticks(['Date'])
#plt.yticks(['Total Traveler Throughput 2020'])

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
