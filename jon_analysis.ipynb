```python
best_correlation=0
correlations= []
case_day_shift= []
counter=0

for i in correlation_total_data:
    if counter<2:
        counter+=1
        continue
    current_corr = correlation_total_data["Total Traveler Throughput 2020"].corr(correlation_total_data[i])
    if (current_corr * current_corr) > (best_correlation*best_correlation):
        best_correlation=current_corr
    correlations.append(current_corr)
    case_day_shift.append(i)
    
```
