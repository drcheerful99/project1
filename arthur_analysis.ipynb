```python
# import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# import dataFrame
from final_data_prep_py import *
```

```python
# reformat date for plot
# nyt_nationwide['Date'] = pd.to_datetime(nyt_nationwide['Date'], format = '%m/%d/%Y')

nyt_nationwide['Date'] = pd.to_datetime(nyt_nationwide['Date'], format='%Y%m%d')
nyt_nationwide.head()
```

```python
# plot total cases change over time 
# create lists for x_axis (date) and y_axis(total cases)
x = nyt_nationwide['Date'].tolist()
y = nyt_nationwide['Total Cases'].tolist()
y_change = nyt_nationwide['Case Rate of Change'].tolist()

plt.figure(figsize = (14, 12))
plt.plot(x, y, color = 'blue', label = 'Total Cases through July, 2020', linewidth = 1)
plt.ticklabel_format(axis="y", style="plain")
plt.title('Total Cases Over Time', fontsize = 24, pad = 20)
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(alpha = 0.75)
plt.legend(loc = 'best')
plt.tight_layout()
plt.show()
```

```python
# plot case rate changes  over time 
# create lists for x_axis (date) and y_axis(total cases)
x = nyt_nationwide['Date'].tolist()
# y = nyt_nationwide['Total Cases'].tolist()
y_change = nyt_nationwide['Case Rate of Change'].tolist()

plt.figure(figsize = (14, 12))
plt.plot(x, y_change, color = 'blue', label = 'Case Rate of Changes', linewidth = 1)
plt.title('Case Rate of Change Over Time', fontsize = 24, pad = 20)
plt.xlabel('Date')
plt.ylabel('Case Rate')
plt.legend(loc = 'best')
plt.grid(alpha = 0.75)
plt.show()
```
